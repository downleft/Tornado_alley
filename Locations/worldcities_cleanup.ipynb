```python
# Dependencies and Setup
import pandas as pd
```

The initial worldcities.csv file accessed in the next cell has come from the World Cities Database that was put together through SimpleMaps and has been accessed using the following link, https://simplemaps.com/data/world-cities, and was accessed on July 3, 2024.

```python
# World Cities Data Files
world_cities_path = "worldcities.csv"

# Read the World Cities data
world_cities_df = pd.read_csv(world_cities_path)
world_cities_df.head()
```

```python
# Sift down to US Cities
us_cities_df = world_cities_df.loc[world_cities_df["country"] == "United States"]
us_cities_df.head()
```

```python
# Drop Extraneous Columns
us_cities_df = us_cities_df.reset_index()
us_cities_df = us_cities_df.drop(columns = ["city_ascii", "country", "iso2", "iso3",  
                                            "capital", "population", "id", "index"])
us_cities_df.head()
```

```python
us_cities_df = us_cities_df.rename(columns = {"admin_name":"state"})
us_cities_df.head()
```

```python
# Export us_cities to CSV file
us_cities_df.to_csv("major_us_cities.csv")
```

```python
# jsonify us_cities
us_cities_json = us_cities_df.to_json("major_us_cites.json", orient = "records")
us_cities_js = us_cities_df.to_json("major_us_cites.js", orient = "records")
```
