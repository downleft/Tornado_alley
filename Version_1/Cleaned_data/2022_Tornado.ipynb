```python
import pandas as pd
import datetime
import pprint
```

```python
df = pd.read_csv("Resources/StormEvents_details-ftp_v1.0_d2022_c20240216.csv")
```

```python
# Checking the data
df.info()
```

```python
# Checking the data
df.head()
```

```python
# Creating a new DataFrame with the states we want to include
state_names = ["KANSAS", "NEBRASKA", "IOWA", "OKLAHOMA", "KENTUCKY",
                     "ALABAMA", "MISSISSIPPI", "TENNESSEE", "LOUISIANA"]

# Filter states we want to include
states_df = df[df["STATE"].isin(state_names)]

# Display DataFrame
states_df.head()
```

```python
#Creating a new DataFrame of columns we want to drop
columns_drop = ["BEGIN_YEARMONTH", "BEGIN_DAY","BEGIN_TIME", "END_YEARMONTH", "END_DAY",
                "END_TIME", "EPISODE_ID", "EVENT_ID", "STATE_FIPS", 
                "END_RANGE", "END_AZIMUTH", "EPISODE_NARRATIVE", "EVENT_NARRATIVE",
                "DATA_SOURCE", "CZ_TYPE", "CZ_FIPS", "CZ_NAME", "TOR_OTHER_CZ_FIPS", 
                "TOR_OTHER_CZ_NAME", "BEGIN_RANGE", "BEGIN_AZIMUTH", "BEGIN_LOCATION",
                "WFO", "CZ_TIMEZONE", "TOR_OTHER_WFO", "TOR_OTHER_CZ_STATE",
                "END_LOCATION", "SOURCE", "INJURIES_INDIRECT", "DEATHS_INDIRECT", 
                "DAMAGE_CROPS", "MAGNITUDE", "MAGNITUDE_TYPE", "FLOOD_CAUSE", "CATEGORY",
                "END_DATE_TIME", "MONTH_NAME", "TOR_LENGTH"]

#Dropping the columns
states_df.drop(columns=columns_drop, inplace=True)

#Display DataFrame
states_df.head()
```

```python
states_df.info()
```

```python
# Filter out rows for only Tornado
tornado_df = states_df[states_df["EVENT_TYPE"] == "Tornado"]

# Display DataFrame
tornado_df.head()
```

```python
# Create a new DataFrame
new_tornado_df = pd.DataFrame()

# Changing the names of each column and arranging them
column_names = {
    "YEAR": "Year",
    "BEGIN_DATE_TIME": "Date",
    "STATE": "State",
    "TOR_F_SCALE": "Rating",
    "INJURIES_DIRECT": "Injuries",
    "DEATHS_DIRECT": "Death",
    "BEGIN_LAT": "Start_Lat",
    "BEGIN_LON": "Start_Lon",
    "END_LAT": "End_Lat",
    "END_LON": "End_Lon",
    "TOR_WIDTH": "Width" 
}

# Reorder columns and rename them
new_tornado_df = tornado_df[column_names.keys()].rename(columns=column_names)

# Display DataFrame
new_tornado_df.head()
```

```python
#Changing Date into datetime format
new_tornado_df["Date"] = pd.to_datetime(new_tornado_df["Date"], format="%m/%d/%y %H:%M")

# Extract only the date part
new_tornado_df["Date"] = new_tornado_df["Date"].dt.date

#Changing Date into datetime format again
new_tornado_df["Date"] = pd.to_datetime(new_tornado_df["Date"])

# Convert latitude, longitude and Width columns to integer
new_tornado_df[["Start_Lat", "Start_Lon", "End_Lat", "End_Lon", "Width"]] = \
new_tornado_df[["Start_Lat","Start_Lon", "End_Lat", "End_Lon", "Width"]].astype("int")

new_tornado_df.info()
```

```python
# Changing the state names to their abbreviations
state_abbr = {
    "KANSAS": "KS",
    "NEBRASKA": "NE",
    "IOWA": "IA",
    "OKLAHOMA": "OK",
    "KENTUCKY": "KY",
    "ALABAMA": "AL",
    "MISSISSIPPI": "MS",
    "TENNESSEE": "TN",
    "LOUISIANA": "LA"
}

# Replaceing state names with the abbreviations
new_tornado_df["State"] = new_tornado_df["State"].str.upper().map(state_abbr)

# Display DataFrame
new_tornado_df.head()
```

```python
new_tornado_df.index.name = "index_name"
new_tornado_df.head()
```

```python
#Removing EF from dataframe
new_tornado_df["Rating"] = new_tornado_df["Rating"].str.replace("EF", " ")

# Display DataFrame
new_tornado_df.head()
```

```python
print(new_tornado_df["Rating"].unique())
```

```python
# Count the number of U 
count_U = (new_tornado_df["Rating"] == "U").sum()
# Count the number of U after stripping whitespaces
count_U = (new_tornado_df['Rating'].str.strip() == "U").sum()

# Display the count
count_U
```

```python
# Filter out rows with U after stripping whitespace
new_tornado_df = new_tornado_df[new_tornado_df["Rating"].str.strip() != "U"]

# Display DataFrame
new_tornado_df.head()
```

```python
# Count the number of U
count_U = (new_tornado_df["Rating"] == "U").sum()
# Count the number of U after stripping whitespaces
count_U = (new_tornado_df["Rating"].str.strip() == "U").sum()

# Display the count
count_U
```
