```python
import pandas as pd
import datetime
import pprint
```

```python
# Read each CSV into a DataFrame
tornado_data = pd.read_csv("Resources/Tornado_data.csv")
data_2022 = pd.read_csv("Resources/2022_data.csv")
data_2023 = pd.read_csv("Resources/2023_data.csv")
```

```python
# Concatenate the Dataframes
merged_df = pd.concat([tornado_data, data_2022, data_2023])

# Display Dataframe
merged_df.head()
```

```python
# Filter the DataFrame
specific_year_df = merged_df[merged_df["Year"] == 2022]

# Display DataFrame
specific_year_df.head()
```

```python
# Filter the DataFrame
specific_year_df = merged_df[merged_df["Year"] == 2023]

# Display DataFrame
specific_year_df.head()
```

```python
# Write the data to a CSV
merged_df.to_csv("merged_data.csv")
```
